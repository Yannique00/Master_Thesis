```python
import pandas as pd
import numpy as np

# # Assume your dataframe is called df and has these columns: ['Patient_ID', 'Datum', 'Hemoglobine']
# df['Datum'] = pd.to_datetime(df['Datum'])
# df = df.sort_values(['Patient_ID', 'Datum'])



# For each patient, slide a 7-day window and find max/min Hb
result = []

for patient_id, group in df_merged4.groupby('pseudo_id'):
    group = group.sort_values('lab_measurement_start_date').reset_index(drop=True)
    
    for i, row in group.iterrows():
        window_start = row['lab_measurement_start_date'] - pd.Timedelta(days=7)
        window_data = group[(group['lab_measurement_start_date'] >= window_start) & (group['lab_measurement_start_date'] <= row['lab_measurement_start_date'])]
        
        if len(window_data) >= 2:
            max_hb = window_data['Hemoglobine'].max()
            min_hb = window_data['Hemoglobine'].min()
            drop_pct = ((max_hb - min_hb) / max_hb) * 100
            
            if drop_pct >= 6.85 and min_hb < 7.5:
                result.append({
                    'pseudo_id': patient_id,
                    'Window_End': row['lab_measurement_start_date'],
                    'lab_measurement_start_date': row['lab_measurement_start_date'],
                    'Hb_Drop_%': drop_pct,
                    'Hb_bool': True if drop_pct >= 6.85 else False,
                    'Max_Hb': max_hb,
                    'Min_Hb': min_hb
                })

df_hb_drops = pd.DataFrame(result)
df_hb_drops.head()
```

```python
def label_lab_values(row):
    label = {}

    if row['lab_measurement_description'] == 'Hemoglobine':
        label['low_hb'] = row['lab_measurement_test_value_numeric'] < 7.5
    elif row['lab_measurement_description'] == 'Hematocriet':
        label['low_hematocriet'] = row['lab_measurement_test_value_numeric'] < 0.35
    elif row['lab_measurement_description'] == 'MCV':
        label['low_mcv'] = row['lab_measurement_test_value_numeric'] < 80
        label['high_mcv'] = row['lab_measurement_test_value_numeric'] > 100
    elif row['lab_measurement_description'] == 'INR':
        label['high_inr'] = row['lab_measurement_test_value_numeric'] > 1.5
    elif row['lab_measurement_description'] == 'APTT':
        label['high_aptt'] = row['lab_measurement_test_value_numeric'] > 35
    elif row['lab_measurement_description'] == 'Protrombinetijd':
        label['high_pt'] = row['lab_measurement_test_value_numeric'] > 15
    elif row['lab_measurement_description'] == 'Ureum':
        label['high_ureum'] = row['lab_measurement_test_value_numeric'] > 7.5
    elif row['lab_measurement_description'] == 'Trombocyten':
        label['low_plt'] = row['lab_measurement_test_value_numeric'] < 150

    return pd.Series(label)
```

```python
# Only keep relevant tests
relevant_tests = [
    'Hemoglobine', 'Hematocriet', 'MCV', 'INR',
    'APTT', 'Protrombinetijd', 'Ureum', 'Trombocyten'
]

df_relevant = df_hb[df_hb['lab_measurement_description'].isin(relevant_tests)].copy()

# Apply the flag function
flagged = df_relevant.apply(label_lab_values, axis=1)

# Combine with pseudo_id
flagged['pseudo_id'] = df_relevant['pseudo_id']
```
