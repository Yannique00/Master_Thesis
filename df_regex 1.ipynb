```python
def build_pattern(term_list):
    return re.compile(r"(" + "|".join(term_list) + r")", flags=re.IGNORECASE)

def extract_matches_and_context(text, pattern, window=5):
    text = str(text)
    matches = []
    for match in pattern.finditer(text):
        print("\u2705 Match found:", match.group())  # DEBUG LINE
        start, end = match.span()
        before = text[:start].split()[-window:]
        after = text[end:].split()[:window]
        context = " ".join(before + [match.group()] + after)
        matches.append(context)
    return matches if matches else None

def extract_match(text, pattern):
    match = pattern.search(text.lower())
    return match.group() if match else None

def apply_term_matching_with_specialism(df, text_col, spec_col, termlist_general, termlist_mdl, target_specialism, prefix):
    df[text_col] = df[text_col].fillna("").astype(str)

    pattern_general = build_pattern(termlist_general)
    pattern_mdl = build_pattern(termlist_mdl)

    def match_fn(row):
        pattern = pattern_mdl if row[spec_col] == target_specialism else pattern_general
        return extract_matches_and_context(row[text_col], pattern)
    
    def match_fn_simple(row):
        pattern = pattern_mdl if row[spec_col] == target_specialism else pattern_general
        return extract_match(row[text_col], pattern)
    
    df[f'{prefix}_match'] = df.apply(match_fn_simple, axis=1)
    df[f'{prefix}_match_2'] = df.apply(match_fn, axis=1)
    df[f'{prefix}_context'] = df.apply(match_fn, axis=1)
    df[f'has_{prefix}'] = df[f'{prefix}_context'].apply(lambda x: x is not None)
    return df
```

```python
def extract_context(text, pattern, window=5):
    text = text.lower()
    matches = []
    for match in pattern.finditer(text):
        start, end = match.start(), match.end()
        before = text[:start]
        after = text[end:]
        before_words = re.findall(r'\w+', before)[-window:]
        after_words = re.findall(r'\w+', after)[:window]
        context = ' '.join(before_words + [match.group()] + after_words)
        matches.append(context)
    return matches[0] if matches else None

def extract_match(text, pattern):
    match = pattern.search(text.lower())
    return match.group() if match else None
```

```python
def find_all_matches(text, pattern):
    """Return all matched terms in a text."""
    return [m.group() for m in pattern.finditer(text)]

def is_negated_report(text, match_pattern, negation_pattern):
    """Return True only if all matches are negated."""
    text = str(text)
    matches = find_all_matches(text, match_pattern)
    if not matches:
        return False  # No matches, so nothing to negate

    negated_terms = find_all_matches(text, negation_pattern)

    # Count how many matched terms appear in a negated form
    negated_count = 0
    for term in matches:
        if any(term.lower() in neg.lower() for neg in negated_terms):
            negated_count += 1

    return negated_count == len(matches)
```

```python
def is_final_match(row):
    # Skip if no AGIB or GIB term matched
    if not row.get('has_agib', False) and not row.get('has_gib', False):
        return False

    # If there's no negation, it's definitely a valid positive
    if not row.get('negated_context'):
        return True

    # Extract all matched terms (agib + gib)
    matched_terms = []

    if isinstance(row.get('agib_context'), list):
        matched_terms += [term.lower() for term in row['agib_context']]
    if isinstance(row.get('gib_context'), list):
        matched_terms += [term.lower() for term in row['gib_context']]

    if not matched_terms:
        return False  # No extractable match text

    # Extract all negated matches (as strings)
    negated_terms = []
    if isinstance(row.get('negated_context'), list):
        negated_terms += [term.lower() for term in row['negated_context']]

    # Check: Are all matched terms negated?
    all_negated = True
    for match in matched_terms:
        if not any(match in neg for neg in negated_terms):
            all_negated = False
            break

    return not all_negated  # True if at least one match is NOT negated
```
