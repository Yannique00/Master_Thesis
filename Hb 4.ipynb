```python
df_tests_wide = df_tests_filtered.pivot_table(
    index=['pseudo_id', 'lab_measurement_start_date'],
    columns='lab_measurement_description',
    values='lab_measurement_test_value_numeric',
    aggfunc='first'
).reset_index()

df_tests_wide
```

```python
## algemeen
df_tests_wide['low_hb'] = df_tests_wide['Hemoglobine'] < 7.5
df_tests_wide['low_hematocriet'] = df_tests_wide['Hematocriet'] < 0.35
#bloedgroep
df_tests_wide['high_ureum'] = df_tests_wide['Ureum'] > 7.5 # voor hoge bloeding 

df_tests_wide['low_mcv'] = df_tests_wide['MCV'] < 80 # voor chronische anemie
df_tests_wide['high_mcv'] = df_tests_wide['MCV'] > 80 # voor eventueel acuut bloedverlies (aangezien waarde dan vaak nog normaal is)


df_tests_wide['high_inr'] = df_tests_wide['INR'] > 1.5 # in geval van coumaringegebruik
df_tests_wide['high_aptt'] = df_tests_wide['APTT'] > 35 # bij leverlijden

df_tests_wide['high_pt'] = df_tests_wide['Protrombinetijd'] > 15
df_tests_wide['high_ureum'] = df_tests_wide['Ureum'] > 7.5
df_tests_wide['low_plt'] = df_tests_wide['Trombocyten'] < 150

# df_tests_wide['high_bilirubine'] = df_tests_wide['Bilirubine totaal'] >= 17

df_tests_wide
```

```python
# Combine features: 1) Hb + Urea  OR  2 or more coagulopathy markers
df_tests_wide['likely_GIB'] = (
    (df_tests_wide['low_hb'] & df_tests_wide['high_ureum']) |
    (df_tests_wide[['low_hb', 'high_inr', 'low_plt', 'high_aptt', 'high_pt']].sum(axis=1) >= 2)
)

print(df_tests_wide.value_counts('likely_GIB'))
df_tests_wide

df_tests_wide['likely_gib'] = (
    df_tests_wide[['low_hb', 'low_hematocriet', 'high_ureum', 'high_inr', 'high_aptt', 'high_pt', 'low_plt']]
    .sum(axis=1) >= 2
)

df_tests_wide['likely_gib'].value_counts()
```

```python
df_tests_wide['likely_agib'] = (
    df_tests_wide['low_hb'] &
    df_tests_wide['high_ureum'] &
    (df_tests_wide['low_mcv'] == False)  # dus MCV normaal of hoog
)

df_tests_wide['likely_agib'].value_counts()
```
