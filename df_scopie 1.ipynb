```python

# Create a DataFrame of SEH visits
seh_visits = df_1[df_1['is_seh']].copy()
seh_visits = seh_visits[['pseudo_id', 'verslagen_report_start_date']]
seh_visits.rename(columns={'verslagen_report_start_date': 'seh_time'}, inplace=True)

# Create a DataFrame of scopie procedures
scopies = df_1[df_1['is_scopie']].copy()
scopies = scopies[['pseudo_id', 'verslagen_report_start_date']]
scopies.rename(columns={'verslagen_report_start_date': 'scopie_time'}, inplace=True)

# Merge on pseudo_id to match patient
merged = seh_visits.merge(scopies, on='pseudo_id', how='inner')

# Compute time delta
merged['time_diff'] = merged['scopie_time'] - merged['seh_time']

# Keep only positive diffs within 24 hours
within_24h = merged[
    (merged['time_diff'] >= timedelta(0)) &
    (merged['time_diff'] <= timedelta(hours=24))
]

within_24h = within_24h.reset_index(drop=True)
within_24h
```
