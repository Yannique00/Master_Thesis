```python
import pandas as pd
import re
from datetime import timedelta
```

```python
df_reports = pd.read_csv('a:/bloeding-met-patientenlijst-gedetailleerd/bloeding-met-patientenlijst-4-verslagen.csv')
# df_reports = pd.read_csv('a:/bloeding-met-patientenlijst-gedetailleerd/verslagen-abb-corrected.csv')


# df_reports['alltext'] = df_reports['verslagen_report_content']
df_reports['verslagen_report_start_date'] = pd.to_datetime(df_reports['verslagen_report_start_date'], errors='coerce')

df_reports.head()
```

```python
df_dbc = pd.read_csv('a:/bloeding-met-patientenlijst-gedetailleerd/bloeding-met-patientenlijst-2-diagnose-db-cs.csv')
df_dbc['diagnose_db_cs_dbc_start_date'] = pd.to_datetime(df_dbc['diagnose_db_cs_dbc_start_date'])
df_dbc['diagnose_db_cs_dbc_end_date'] = pd.to_datetime(df_dbc['diagnose_db_cs_dbc_end_date'])
df_dbc.head()
```

```python
df_afspraken = pd.read_csv('a:/bloeding-met-patientenlijst-gedetailleerd/bloeding-met-patientenlijst-5-afspraken.csv')
df_afspraken['afspraken_appointment_start_date'] = pd.to_datetime(df_afspraken['afspraken_appointment_start_date'])

df_afspraken.head()
```

```python
for i in range(len(df_reports)):
    for j in range(len(df_dbc)):
        if df_reports['pseudo_id'][i] == df_dbc['pseudo_id'][j]:
            date = df_reports['verslagen_report_start_date'][i]

            if df_dbc['diagnose_db_cs_dbc_start_date'][j] <= date <= df_dbc['diagnose_db_cs_dbc_end_date'][j]:
                df_reports.at[i, 'diagnose_db_cs_dbc_specialty_code'] = df_dbc['diagnose_db_cs_dbc_specialty_code'][j]
                df_reports.at[i, 'diagnose_db_cs_dbc_diagnosis_code'] = df_dbc['diagnose_db_cs_dbc_diagnosis_code'][j]
                df_reports.at[i, 'diagnose_db_cs_dbc_start_date'] = df_dbc['diagnose_db_cs_dbc_start_date'][j]
                df_reports.at[i, 'diagnose_db_cs_dbc_end_date'] = df_dbc['diagnose_db_cs_dbc_end_date'][j]

df_reports
```

```python
# Create empty columns for results if they don't exist yet
df_reports['afspraken_appointment_description'] = None
df_reports['afspraken_appointment_specialism'] = None

for i in range(len(df_reports)):
    pseudo_id = df_reports.loc[i, 'pseudo_id']
    report_datetime = df_reports.loc[i, 'verslagen_report_start_date']
    report_date = report_datetime.date()

    # Filter potential matches (same pseudo_id and same calendar day)
    potential_matches = df_afspraken[
        (df_afspraken['pseudo_id'] == pseudo_id) &
        (df_afspraken['afspraken_appointment_start_date'].dt.date == report_date)
    ]

    if not potential_matches.empty:
        # Calculate absolute time difference
        potential_matches['time_diff'] = (potential_matches['afspraken_appointment_start_date'] - report_datetime).abs()

        # Find the closest appointment in time
        closest_idx = potential_matches['time_diff'].idxmin()
        df_reports.at[i, 'afspraken_appointment_description'] = df_afspraken.loc[closest_idx, 'afspraken_appointment_description']
        df_reports.at[i, 'afspraken_appointment_specialism'] = df_afspraken.loc[closest_idx, 'afspraken_appointment_specialism']
        # df_reports.at[i, 'afspraken_appointment_start_date'] = df_afspraken.loc[closest_idx, 'afspraken_appointment_start_date']

df_reports.head()
```

```python
df = df_reports
```

```python
df_1 = df[['pseudo_id', 'verslagen_report_start_date', 'verslagen_report_content', 
                 'diagnose_db_cs_dbc_specialty_code', 'diagnose_db_cs_dbc_diagnosis_code', 
                 'afspraken_appointment_description', 'afspraken_appointment_specialism',
                 'verslagen_report_specialism', 'diagnose_db_cs_dbc_start_date',
                    'diagnose_db_cs_dbc_end_date'
                 ]].copy()
```

```python
df_1['is_seh'] = (
    df_1['afspraken_appointment_description'].str.contains('SEH', case=False, na=False) |
    (df_1['afspraken_appointment_specialism'] == 'Spoedeisende Hulp') |
    (df_1['verslagen_report_specialism'] == 'Spoedeisende Hulp')
)
df_1.head()
```

```python
scopie_keywords = ['scopie', 'ercp', 'video', 'endoscopie', 'gastroscopie', 'coloscopie', 'endoscopisch']

df_1['is_scopie'] = (
    df_1['verslagen_report_specialism'].str.lower().str.contains('|'.join(scopie_keywords), na=False) |
    df_1['afspraken_appointment_description'].str.lower().str.contains('|'.join(scopie_keywords), na=False)
)
df_1.head()
```

```python

# Create a DataFrame of SEH visits
seh_visits = df_1[df_1['is_seh']].copy()
seh_visits = seh_visits[['pseudo_id', 'verslagen_report_start_date']]
seh_visits.rename(columns={'verslagen_report_start_date': 'seh_time'}, inplace=True)

# Create a DataFrame of scopie procedures
scopies = df_1[df_1['is_scopie']].copy()
scopies = scopies[['pseudo_id', 'verslagen_report_start_date']]
scopies.rename(columns={'verslagen_report_start_date': 'scopie_time'}, inplace=True)

# Merge on pseudo_id to match patient
merged = seh_visits.merge(scopies, on='pseudo_id', how='inner')

# Compute time delta
merged['time_diff'] = merged['scopie_time'] - merged['seh_time']

# Keep only positive diffs within 24 hours
within_24h = merged[
    (merged['time_diff'] >= timedelta(0)) &
    (merged['time_diff'] <= timedelta(hours=24))
]

within_24h = within_24h.reset_index(drop=True)
within_24h
```

```python
df_1['scopie_after_seh'] = None
df_1['scopie_time'] = None
df_1['time_diff'] = None


for i in range(len(df_1)):
    for j in range(len(within_24h)):
        if df_1['pseudo_id'][i] == within_24h['pseudo_id'][j]:
            if df_1['verslagen_report_start_date'][i] == within_24h['seh_time'][j]:
                df_1.at[i, 'scopie_after_seh'] = True
                df_1.at[i, 'scopie_time'] = within_24h['scopie_time'][j]
                df_1.at[i, 'time_diff'] = within_24h['time_diff'][j]
                break
            else:
                df_1.at[i, 'scopie_after_seh'] = False
```

```python
df_scopie = df_1[df_1['scopie_after_seh'] == True]
df_scopie.head()
```

```python
scopie_patient_lst = df_scopie['pseudo_id'].unique().tolist()
len(scopie_patient_lst)
```
